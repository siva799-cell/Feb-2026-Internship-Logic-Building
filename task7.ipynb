```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # default status

    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is {self.status}")


# Taking input
light_name = input("Light Name: ")
action = input("Action: ").upper()

# Create object
light = SmartLight(light_name)

# Perform action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action")
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Taking input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Creating object
emp = Employee(name, emp_id, department)

# Display ID card
emp.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Taking input
name = input("Contact Name: ")
phone = input("Phone Number: ")

# Creating object
c = Contact(name, phone)

# Display contact details
c.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + self.price)


# Taking input
name = input("Product Name: ")
price = input("Price: ")

# Creating object
p = Product(name, price)

# Display price tag
p.display_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie Name:", self.name)
        print("Rating:", self.rating)


# Taking input
name = input("Movie Name: ")
rating = input("Rating: ")

# Creating object
m = Movie(name, rating)

# Display movie details
m.display_movie()
```

```python
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)


# Taking input
name = input("Customer Name: ")
address = input("Address: ")

# Creating object
d = Delivery(name, address)

# Display delivery details
d.display_delivery()
```
