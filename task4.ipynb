```python
def recharge():
    amounts = [199,299,399,599]
    while True:
        amount = int(input("Enter Amount to recharge"))
        if amount < 50:
            print("Invalid Enter Amount Greater than or equals to 50")
            continue 
        else:
            if amount in amounts:
                print(f"{amount} Recharged SuccessFully")
                break
recharge()                
```

```python
def check_stock(products):
    for item, quantity in products.items():
        if quantity < 15:
            print(f"{item}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{item}: Stock OK (Stock = {quantity})")


# Example dictionary
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Salt": 18
}

check_stock(inventory)
```

```python
def marks():
    l = list(map(int, input("Enter numbers separated by space: ").split()))
    a = sum(l)
    b = len(l)
    c = a/b
    if c >= 50:
        print("Pass")
    else:
        print("Fail")
marks()    
```

```python
def cab_fare():
    BASE_FARE = 50
    RATE_PER_KM = 12

    while True:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        fare = BASE_FARE + (km * RATE_PER_KM)

        if peak == "yes":
            fare = fare * 1.25   # 25% extra

        print(f"Total Fare = â‚¹{fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Calculator ")
            break

cab_fare()
```

```python
def employee():
    l = list(input("Enter P/A separated by space: ").split())
    p_count = l.count('p')
    length = len(l)
    avg = (p_count/length)*100
    if avg >= 75:
        print("Eligible")
    else:
        print("Not Eligible")
employee()
```

```python
def check_password(password):
    if len(password) < 8:
        print("Weak Password (Minimum length should be 8)")
        return

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password (Must include digit and special character @#$)")


# input
pwd = input("Enter password: ")
check_password(pwd)
```
