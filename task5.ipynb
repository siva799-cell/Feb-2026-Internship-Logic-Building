```python
#prob3
def adequacy(rainfall,required):
    total = 0
    for i in rainfall:
        total+=i
    average = total/len(rainfall)
    if average >= required:
        print("Adequate Rainfall")
    else:
        print("Non -Adequate Rain fall")
rainfall = list(map(int,input("Enter Data").split()))
required = int(input("Enter The required rainfall"))
adequacy(rainfall,required)
```

```python
def check_duplicate_usernames(usernames):
    unique_users = set()   
    duplicate_found = False
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
user_list = input("Enter usernames separated by spaces: ").split()

check_duplicate_usernames(user_list)
```

```python
def application_health(error_count):   
    print("Error Count:", error_count)
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
errors = int(input("Enter error count: "))

application_health(errors)
```

```python
def validate_transaction(amount):
    daily_limit = 50000    
    print("Transaction Amount:", amount)    
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")
transaction_amount = int(input("Enter transaction amount: "))
validate_transaction(transaction_amount)
```

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for day in attendance_list:
        if day == 1:   # 1 means Present, 0 means Absent
            present_days += 1
    attendance_percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", attendance_percentage)
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance = list(map(int, input("Enter attendance (1=Present, 0=Absent) separated by spaces: ").split()))
check_exam_eligibility(attendance)
```

```python
def calculate_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
units_consumed = int(input("Enter units consumed: "))
total_bill, usage_status = calculate_electricity_bill(units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
